Z score method to find outliers

```python
data = df['Age']
z_scores =(data - data.mean())/data.std()
```

```python
z_scores.mean()
```

```python
threshold = 2
```

```python
outliers = df[abs(z_scores)> threshold]
print('outliers:')
outliers
```
